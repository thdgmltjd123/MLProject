```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

columns = ['label']
for i in range(1,785):
  columns.append(i)
columns

data = pd.read_csv("/content/drive/My Drive/Colab Notebooks/mnist_test.csv", names = columns)
```

```
data.head()
```

# **Plot the average image**

```
li = []
count = 0
plt.figure(figsize = (12,5))
for i in range(10):
  count += 1
  for j in range(1,785):
    li.append(data.loc[data.label == i].iloc[:,j].mean())
  plt.subplot(2, 5, count)
  plt.imshow(np.array(li).reshape(28,28),cmap = 'Greys')
  plt.title(i)
  li = []
plt.show()
```
