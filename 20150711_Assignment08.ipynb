# **Plot the average image**

```
li = []
count = 0
plt.figure(figsize = (12,5))
for i in range(10):
  count += 1
  for j in range(1,785):
    li.append(data.loc[data.label == i].iloc[:,j].mean())
  plt.subplot(2, 5, count)
  plt.imshow(np.array(li).reshape(28,28),cmap = 'Greys')
  plt.title(i)
  li = []
plt.show()
```

# **Present the output of the neural network with random weight**

```
import random

theta = []
for i in range(784):
  theta.append(random.gauss(0.0,1.0))
theta = np.array(theta).reshape(-1,1)
```

```
def sigmoid(arr):
  return 1 / (1 + np.exp(-arr))
```

```
X = np.array(data.iloc[:,1:])

logit_value = sigmoid(np.dot(X,theta))
data["logit"] = logit_value
```

```
for i in range(10):
  print("라벨 {}의 sigmoid 값 평균 : {}".format(i,data.loc[data.label == i].iloc[:,785].mean()))
```
