# **데이터 불러오기**

```python
import pandas as pd
import numpy as np

train_data = pd.read_csv("/content/drive/My Drive/Colab Notebooks/data_train.csv",names = ['x','y','z','h'])
test_data = pd.read_csv("/content/drive/My Drive/Colab Notebooks/data_test.csv" , names = ['x','y','z','h'])
```

# **세타값 초기화 감잡기 plot**

```python
plt.figure(figsize=(5,5))
plt.plot(train_data.iloc[:,3])
```

```python
#세타 값 초기화
theta0 = -2
theta1 = 2.5
theta2 = 10
theta3 = -8
```

# Multiple Linear Regression 구현

```python
#훈련 데이터
x = train_data.iloc[:,0]
y = train_data.iloc[:,1]
z = train_data.iloc[:,2]
h = train_data.iloc[:,3]
#테스트 데이터
test_x = test_data.iloc[:,0]
test_y = test_data.iloc[:,1]
test_z = test_data.iloc[:,2]
test_h = test_data.iloc[:,3]

#데이터 길이
length = len(h)
test_length = len(test_h)

#데이터 적합식
fxyz = theta0 + theta1 * x + theta2 * y + theta3 * z
test_fxyz = theta0 + theta1 * test_x + theta2 * test_y + theta3 * test_z

#데이터 loss함수
energy = sum((fxyz - h) ** 2) / (length * 2)
test_energy = sum((test_fxyz - test_h) ** 2) / (test_length * 2)

#훈련용 임시변수
temp_energy = energy

#학습률
alpha = 0.00002

#세타 보관소
theta0_val = [theta0]
theta1_val = [theta1]
theta2_val = [theta2]
theta3_val = [theta3]
energy_val = [energy]
test_energy_val = [test_energy]

while True:    
    gra_t0 = sum(fxyz - h) / length 
    gra_t1 = sum((fxyz - h) * x) / length
    gra_t2 = sum((fxyz - h) * y) / length
    gra_t3 = sum((fxyz - h) * z) / length
    
    theta0 -= alpha * gra_t0
    theta1 -= alpha * gra_t1
    theta2 -= alpha * gra_t2
    theta3 -= alpha * gra_t3
    
    fxyz = theta0 + theta1 * x + theta2 * y + theta3 * z
    #세타가 갱신된 테스트 적합식
    test_fxyz = theta0 + theta1 * test_x + theta2 * test_y + theta3 * test_z
    theta0_val.append(theta0)
    theta1_val.append(theta1)
    theta2_val.append(theta2)
    theta3_val.append(theta3)

    next_energy = sum((fxyz - h) ** 2) / (length * 2)
    #세타가 갱신된 테스트 loss 값
    test_energy = sum((test_fxyz - test_h)**2) / (test_length * 2)
    energy_val.append(next_energy)
    test_energy_val.append(test_energy)
    
    if(temp_energy - next_energy < 0.001):
        break
    temp_energy = next_energy
```

# **결과 값 시각화**

```python
#세타 수렴 시각화
plt.figure(figsize = (8,6))
plt.title("values of theta",fontsize = 20)
plt.ylabel("value",fontsize = 15)
plt.xlabel("iteration",fontsize = 15)
plt.plot(theta0_val, c = "black",linewidth = 3, label = "theta0")
plt.plot(theta1_val, c = "red",linewidth = 3, label = "theta1")
plt.plot(theta2_val, c = "green",linewidth = 3, label = "theta2")
plt.plot(theta3_val, c = "blue",linewidth = 3, label = "theta3")
plt.legend(loc = 'best')
plt.show()
```

```python
#에너지 수렴 시각화
plt.figure(figsize = (8,6))
plt.title("Energy Convergence",fontsize = 20)
plt.ylabel("value",fontsize = 15)
plt.xlabel("iteration",fontsize = 15)
plt.plot(energy_val, c = "blue",linewidth = 3, label = "Energy")
plt.legend(loc = 'best')
plt.show()
```

```python
#테스트 에너지 시각화
plt.figure(figsize = (8,6))
plt.title("Test Energy",fontsize = 20)
plt.ylabel("value",fontsize = 15)
plt.xlabel("iteration",fontsize = 15)
plt.plot(test_energy_val,c = "red",linewidth = 3, label = "Test_Energy")
plt.legend(loc = 'best')
plt.show()
```
